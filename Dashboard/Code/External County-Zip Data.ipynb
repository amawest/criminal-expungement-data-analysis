```python
import pandas as pd
```

```python
county_zips = pd.read_csv('county_zip_raw.csv')
```

```python
cities   = county_zips[county_zips['County'].str.contains("city")]
cities
```

```python
counties =  county_zips[~county_zips['County'].str.contains("city")]
counties['County'] = counties['County'].astype(str) + " County"
counties
```

```python
county_zip_comparisons = pd.concat([cities, counties])
county_zip_comparisons = county_zip_comparisons[['Zip Code', 'County']]
```

```python
county_zip_comparisons.to_csv('county_zip_edited.csv')
```
